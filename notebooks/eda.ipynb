#### Categorical Features Analysis

In this section, we analyze the categorical features in the dataset by visualizing their distributions and examining their relationship with the target variable. For each categorical feature, we plot two visualizations side by side: one showing the **distribution of the feature** itself, and the other displaying **how the target variable is distributed across the categories**. Additionally, we include pie charts to illustrate the **distribution of each categorical feature separately for employees who stayed (No Attrition) and those who left (Yes Attrition)**.

```python
def cat_by_target(col_name: str):
  """
  Visualizes the distribution of a categorical column by `Attrition` column.
  """
  if col_name not in df.columns:
    raise ValueError(f"Column '{col_name}' not found in DataFrame.")  
  
  cross_tab = pd.crosstab(df[col_name], df[target_col], normalize="columns") * 100
  pie_plot = cross_tab.plot(kind="pie", subplots=True, colormap="coolwarm", figsize=(12, 6), 
                            autopct="%1.1f%%", legend=False, fontsize=10, 
                            title=f"{col_name} Distribution Among Non-Attrition(Left) vs. Attrition(Right)", startangle=90)
  pie_plot.ravel()[0].set_ylabel("")
  pie_plot.ravel()[1].set_ylabel("")
  
  plt.tight_layout()
```

```python
for col in cat_cols:
  if col != target_col and col not in one_value_cols: cat_by_target(col)
  fig = viz.plot_dual_distributions(df, target_col=target_col, col_name=col)
  viz.save_figure(fig, f"{col}_dual_distribution", "cat_features_analysis")  # Only saves the last figure
```
